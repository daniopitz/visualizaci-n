# Problema

El archivo 2020-06-01-CasosConfirmados-totalRegional.csv contiene el número de casos confirmados de covid-19 por región del día 2020-06-01 y el archivo pcr.csv contiene el número de tests PCR realizado por región por día. Programe un código en Jupyter que devuelva un archivo nuevo llamado positividad.csv que contenga el número de tests PCR realizados, el número de tests positivos, la positividad (el número de casos positivos por número de tests realizados) para el día 2020-06-01 para cada región. Entregue el Jupyter Notebook con la solución.

# Solución Problema

```python
import pandas as pd
import numpy as np
```

```python
casos=pd.read_csv('2020-06-01-CasosConfirmados-totalRegional.csv') #abrimos el archivo
```

```python
#eliminamos los acentos
casos['Region']=casos['Region'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
casos['Region']=casos['Region'].str.strip() #eliminamos los espacios vacios adelante y atras
casos['Region']=casos['Region'].replace(r'\s+', ' ', regex=True) #eliminamos los espacios dobles o más grandes en medio
```

```python
casos
```

```python
pcr=pd.read_csv('pcr.csv') #abrimos el otro archivo
```

```python
pcr['Region']=pcr['Region'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
pcr['Region']=pcr['Region'].replace(r'\s+', ' ', regex=True)
pcr['Region']=pcr['Region'].str.replace('OHiggins', "O\'Higgins") #reemplazamos el nombre
```

```python
pcr
```

### Merge

```python
casos_final=pd.merge(casos, pcr[['Region','2020-06-01']], on='Region', how='inner')
casos_final
```

```python
casos_final.columns
```

```python
casos_final['positividad']=casos_final['Casos  nuevos  totales']/casos_final['2020-06-01']
casos_final
```

```python
casos_final['positividad']=casos_final['positividad'].replace(np.inf, 0) #reemplazamos inf por 0
```

```python
casos_final
```

```python
#cambiamos el nombre a algunas columnas
casos_final=casos_final.rename(columns={'Casos  nuevos  totales':'nº de tests positivos',  '2020-06-01': 'nº de tests'})
casos_final
```

```python
#guardamos el archivo con las columnas que nos interesan
casos_final[['Region','nº de tests positivos','nº de tests', 'positividad']].to_csv('positividad.csv', sep=';', index=False)
```
